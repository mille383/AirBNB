```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
air_bnb.describe()
```

```python
air_bnb.sort_values(by="price", ascending=False).head(20)
```

```python
# Which hosts are the busiest and why?

busiest_hosts = air_bnb.groupby('host_id').number_of_reviews.sum()
busiest_hosts.sort_values(ascending=False).head(5)

# Grouping the hosts by number of reviews, showing the top 5. 
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnb['neighbourhood_group'].value_counts()

# 5 groups available, Manhattan shows up the most, but close with Brooklyn
```

```python
# Are private rooms the most popular in manhattan?

manhattan_rooms = air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan']
manhattan_rooms['room_type'].value_counts(normalize=True) * 100.

# NO, entire home/apt are the most popular in Manhattan at ~ 61%. Private rooms make up ~37% of total
```

```python
# Which hosts are the busiest and based on their reviews?

# See question 1
```

```python
# Which neighorhood group has the highest average price?

avg_price_neighborhood = air_bnb.groupby('neighbourhood_group').price.mean()
avg_price_neighborhood.sort_values(ascending=False)

# Manhattan does, followed by Brooklyn. Bronx is cheapest
```

```python
# Which neighbor hood group has the highest total price?

total_price_neighborhood = air_bnb.groupby('neighbourhood_group').price.sum()
total_price_neighborhood.sort_values(ascending=False)

# Manhattan. Staten Island is lowest despite middle of average price
```

```python
#Which top 5 hosts have the highest total price?

host_total_price = air_bnb.groupby(["host_id"], as_index=False).price.sum()
host_total_price.sort_values(by = 'price', ascending=False).head(5)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

noavail_100review = air_bnb[(air_bnb['availability_365'] == 0) & (air_bnb['number_of_reviews'] > 100)]
noavail_100review['host_name']

# all of these folks...
```

```python
# What host has the highest total of prices ...?

host_total_price = air_bnb.groupby(["host_id"], as_index=False).price.sum()
host_total_price.sort_values(by='price', ascending=False).head(1)
```

```python
# ... and where are they located?
air_bnb[air_bnb['host_id'] == 219517861].neighbourhood_group

# Manhattan. This guy appears to be running a hotel chain
```

```python
# When did Danielle from Queens last receive a review?

danielle = air_bnb[(air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] == 'Queens')]
danielle.sort_values(by='last_review', ascending=False).head(1)

# 2019-07-08
```
